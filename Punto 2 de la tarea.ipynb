# Algoritmo que apróxima la envolvente cónica convexa de un conjunto finito de puntos.

**Introducción**.

Acontinuación se presenta un algoritmo que apróxima  el cóno convexo de un conjunto de puntos,esto es un algoritmo que da una cantidad tan grande como se quiera de puntos que pertenezan al cono, la idea es la siguiente: Se introducen los puntos para los cuales se quiere hallar el cono (por ejemplo si quiero hallar la envolvente cónica convexa de los puntos (1 0 1 4)  y (2 1 4 3), introduzco en la matriz A los puntos de la siguiente manera A=[1 0 1 4; 2 1 4 3], todos los puntos que escoja deben pertenecer al mismo $\mathbb{R}^n$), además introduzco hasta donde quiero que varien los coeficientes de la combinación cónica (claramente tienen que ser mayoreso iguales a 0), luego lo que hace el algotimo es generar un vector dsp que alvergara coeficientes desde 0 hasta donde se halla querido, distanciados tanto como se quiera, luego se genera una matriz $m$-dimencional (donde $m$ es la cantidad de filas de A), que contendra todas las posibles combinaciones conicas de las filas de A (esto es posible gracias a un doble ciclo for), luego en caso de que que los puntos que se hallan puestos esten en $\mathbb{R}^2$ o $\mathbb{R}^3$ se imprimira la grafica del cono, en caso contrario solo se podra ver la matriz $m-$Dimencional, que contiene puntos del cono. 

```julia
#Aquí se ve un ejemplo con los puntos (1 1 1), (0 1 0), (1 0 0).

A=[1 1 1;0 1 0;1 0 0]
stopp=10
longitud=10
```

```julia
#Vector que contiene los coeficientes desde 0 hasta stopp

dsp=range(0, stop=stopp,length=longitud)
```

```julia

#Matriz m-dimencional "pts" que contiene los coeficientes de la combinación cónica.

for i in 1:size(A)[1]
    
pts=[x*A[1,:]+y*A[2,:]+z*A[3,:] for x in dsp, y in dsp, z in dsp]
    
end


#En esta parte no supe como hacer para  expandir la matriz a una matriz m dimensional (donde m es la cantidad de vectores), puesto
#que no se como ciclar la cantidad de fors en la matriz pts. (hasta que arregle esto solo funciona para 3 vectores)


pts
```

```julia
#Impreción de las gráficas, en caso de que los puntos esten en R^2 o R^3 .


using Random
using Plots



if size(A)[2]==3
for i in 1:longitud, j in 1:longitud, k in 1:longitud, l in 1:size(A)[1]
scatter!([pts[i,j,k][1]], [pts[i,j,k][2]], [pts[i,j,k][3]], color = "green", label = "", markersize = 2);
scatter!([A[l, 1]], [A[l,2]], [A[l,3]], color = "red", label = "", markersize = 4); 
end

scatter!([pts[2,2,2][1]], [pts[2,2,2][2]], [pts[2,2,2][3]], color = "green", label = "", markersize = 2)
    
 
    
elseif size(A)[2]==2
    
    for i in 1:longitud, j in 1:longitud , l in 1:size(A)[1]
scatter!([pts[i,j][1]], [pts[i,j][2]], color = "green", label = "", markersize = 2); 
        scatter!([A[l, 1]], [A[l,2]], color = "red", label = "", markersize = 4);
end
    

scatter([pts[2,2][1]], [pts[2,2][2]], color = "green", label = "", markersize = 2)
    

    else
        println("No hay una representeación geométrica") 
        end
```
